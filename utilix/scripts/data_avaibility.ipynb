```python
def calculate_percentage(df, plugins=['peak_basics', 'event_basics']):
    
    modes = df['mode'].unique()
    percentages = []
    
    for mode in modes:

        mode_df = df[df['mode'] == mode]
        mode_percentages = {}
        mode_percentages['Mode'] = mode
        
        for p in plugins:
            is_stored = np.array([st.is_stored(r, p) for r in mode_df['name']])
            tot_lenght = len(is_stored)
            _true = len(is_stored[is_stored==True])
            mode_percentages[f"{p}"] = f'{_true}/{tot_lenght} ({100*_true/tot_lenght}%)'
        
        percentages.append(mode_percentages)
    
    percentages_df = pd.DataFrame(percentages)
    
    return percentages_df
```

```python
query = {"number": 49918}
rd = coll.find_one(query)

# The output is a dictionary, so we can access the keys and values
for k in rd.keys():
    print(f"{k}")
```

```python
for d in rd['data'][:1]:
    print(f"- {d['type']}")
    print(f"    - Location: {d['location']}")
    print(f"    - Host: {d['host']}")
    print(f"    - Protocol: {d['protocol']}")
    print(f"    - Size: {d.get('size_mb', 10)} MB")
    print(f"    - File count: {d['file_count']}")
    print(f"    - Meta:")
    for k,v in d['meta'].items():
        print(f"        - {k}: {v}")
```

```python
print(f"- Run number: {rd['number']}")
print(f"- Start time: {rd['start']}")
print(f"- End time: {rd['end']}")
print(f"- Duration: {rd['end'] - rd['start']}")

print(f"- Mode: {rd['mode']}")
print(f"- Source: {rd['source']}")

print("- Tags:")
for t in rd['tags']:
    print(f"    {t['name']}")

print("- Comments:")
for c in rd['comments']:
    print(f"    {c['comment']}")

print("- Detectors:")
for d in rd['detectors']:
    print(f"    {d}")

print("- Rates:")
for r,v in rd['rate'].items():
    print(f"    {r:<15}: {v['avg']:.2f} ({v['max']:.2f}) MB/s")

```
